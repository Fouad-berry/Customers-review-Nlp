# 🗣️ 01 — Exploration des avis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style='whitegrid')

df = pd.read_csv('../data/raw/customer_reviews.csv')
print(f'Shape : {df.shape}')
df.head()
```

## Stats descriptives

```python
df['rating'].value_counts().sort_index()
```

```python
df.groupby('company').agg(
    reviews=('review_id', 'count'),
    avg_rating=('rating', 'mean'),
    pct_verified=('verified_purchase', 'mean'),
).round(2)
```

## Exemples d'avis

```python
for r in [1, 3, 5]:
    print(f'\n=== Note {r} ★ ===')
    print(df.loc[df['rating']==r, 'review_text'].iloc[0])
```

## Distribution longueur

```python
fig, ax = plt.subplots(figsize=(10, 5))
df['review_length'].hist(bins=30, ax=ax, edgecolor='black')
ax.set_title('Distribution de la longueur des avis')
ax.set_xlabel('Caractères'); plt.show()
```
