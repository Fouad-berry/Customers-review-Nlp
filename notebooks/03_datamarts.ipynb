# 🧱 03 — Construction des datamarts

```python
import pandas as pd
df = pd.read_csv('../data/processed/reviews_enriched.csv')
print(f'Shape : {df.shape}')
df.head()
```

## DM1 : KPIs globaux

```python
kpi = pd.DataFrame([{
    'total_reviews': len(df),
    'pct_positive': round(100 * (df['sentiment_label']=='positive').mean(), 2),
    'pct_negative': round(100 * (df['sentiment_label']=='negative').mean(), 2),
    'avg_rating': round(df['rating'].mean(), 2),
}])
kpi
```

## DM2 : Performance par entreprise

```python
df.groupby('company').agg(
    reviews=('review_id', 'count'),
    avg_rating=('rating', 'mean'),
    pct_negative=('sentiment_label', lambda s: 100*(s=='negative').mean()),
).round(2).sort_values('pct_negative', ascending=False)
```

## DM3 : Topic × sentiment

```python
df.groupby(['top_topic', 'sentiment_label']).size().unstack(fill_value=0)
```
